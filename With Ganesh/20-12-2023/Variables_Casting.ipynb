```python
#20-12-23
"""
Casting is specifing the data type of a variable.
This can be done by :
"""
String = str(123)# X will be 123 but of format string - Str  the short form for string
Integer = int(321) # Y will be 123". int is the short form for Integer. You Cannot do int("text") because it simply dosen't make sense and python has restricted it
FloatingPoint1 = float(3)#If this is printed (Refer below) It'll show as 3.0 and not only 3 because float meeans that there is a Decimal Point. The full form of float is floating point
FloatingPoint2 = float(1.2) #If this is printed it'll say 1.2 only and not 1.2.0 because there is a decimal point in the (1.2). The fill form of float is floating point


print(String)
print(Integer)
print(FloatingPoint1)
print(FloatingPoint2)


#----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Now, let's ask python to check the Datatypes of the variables, to do that we need to :

print(type(String))
print(type(Integer))
print(type(FloatingPoint1))
print(type(FloatingPoint2))




```
