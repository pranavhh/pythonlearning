**Multiline Strings:**
Strings which are in 2 or more lines.Made by two ways - Starting with a " and ending with the same as well as Starting and ending with '                                                                                                
Let me show you a few examples!(One with Quotation Mark and One with apostrophy)





```python
FiveFacts='''1. American Flags left on the moon will be bleached by the sun
          2. While they are hibernating, bears do not urinate. Their bodies convert waste into protein.
          3. White-faced capuchin monkeys greet each other by sticking their fingers up each others’ noses.
          4. Gummy bears were originally called "dancing bears."
          5. New Zealand has more cats per person than any other country in the world'''

#This has ' mark
print(FiveFacts)
```

```python
FiveNames="""1. Pranav
          2. Ganesh
          3.Hemahlathaa
          4.Harish
          5.Jawahar"""

#This has " mark
print(FiveNames)
```
